# Importing libraries. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.figure_factory as ff
```

# Importing Data sets

```python
df_wine = pd.read_csv("winequality-red.csv")
```

# Data cleansing and Analyzing.

```python
df_wine.shape
```

```python
df_wine.isnull().sum()
```

```python
df_wine.columns=df_wine.columns.str.upper()
```

```python
df_wine.head(10)
```

```python
df_wine.describe()
```

```python
#Quantifying the quality
df_wine['QUALITY_C']=pd.cut(x=df_wine['QUALITY'],bins=[3,5,7,10],labels=['Low','Good','Very Good'],include_lowest=True)
```

```python
df_wine.head()
```

```python
df_wine['QUALITY_C'].unique()
```

```python
df_s=pd.DataFrame(df_wine['QUALITY_C'].value_counts()).reset_index()
```

```python
df_s
```

```python
df_s.columns=['Quality','Count']
```

```python
df_s
```

# Performing EDA

```python
px.bar(data_frame=df_s,x='Quality',y='Count',color='Quality',template='plotly_dark')
```

* This graph represents the value count of the total number of wines that come into their respective category - 
We see that in Good - There are 837 wines , Low = 744 and in very good = 18

```python
#For creating a total summary of the boxplot
plt.figure(figsize=(25,25))
for i,n in enumerate(df_wine.columns[:-1]):
    plt.subplot(3,4,i+1)
    sns.boxplot(data=df_wine,x=n,color='violet')
```

* This toal boxplots represents all the quantitative or numeric columns with the total number of outliers , hence the same has been implemented using the box plot

```python
plt.figure(figsize=(25,25))
for i,n in enumerate(df_wine.columns[:-2]):
    plt.subplot(3,4,i+1)
    sns.histplot(data=df_wine,x=n,kde=True,color='blue')
```
